```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_minus1 = np.array([[1, 2], 
                     [1, 3],
                     [2, 3],
                     [3, 4]])

x_plus1 = np.array([[0, 0.5], 
                     [1, 0],
                     [2, 1],
                     [3, 0]])

fig, ax = plt.subplots()
ax.scatter(x_minus1[:,0], x_minus1[:,1], label='y = -1', color='r')
ax.scatter(x_plus1[:,0], x_plus1[:,1], label='y = 1', color='b')
ax.set_xlabel('X_1')
ax.set_ylabel('X_2')
ax.legend()
```
