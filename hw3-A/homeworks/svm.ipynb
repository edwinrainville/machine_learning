```python
import numpy as np
import matplotlib.pyplot as plt

x_minus1 = np.array([[1, 2], 
                     [1, 3],
                     [2, 3],
                     [3, 4]])

x_plus1 = np.array([[0, 0.5], 
                     [1, 0],
                     [2, 1],
                     [3, 0]])
x1_vals_svm = np.arange(0,4)
weights = [0.4, -1.6]
offset = 1.8
x2_svm = -x1_vals_svm*(weights[0]/weights[1]) - offset/weights[1]

fig, ax = plt.subplots()
ax.scatter(x_minus1[:,0], x_minus1[:,1], label='y = -1', color='r')
ax.scatter(x_plus1[:,0], x_plus1[:,1], label='y = 1', color='b')
ax.plot(x1_vals_svm, x2_svm, color='k', label='Decision Boundary')
ax.set_xlabel('X_1')
ax.set_ylabel('X_2')
ax.legend()
```
